```python
import pandas_datareader as pdr
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# Read in price Data
start = dt.datetime(2021,1,1)
end = dt.datetime.now()

df = pdr.DataReader('ETH-USD', 'yahoo', start, end)
print(df)
```

```python
cols = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']
df = df[cols]
df
```

```python
df.to_csv("./data/ETH-USD_2021-2022.csv")
```
